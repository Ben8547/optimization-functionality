Simulated Annealing example

```python
from simulatedAnnealing import simulated_annealing as sa
import numpy as np
```

```python
def paraboloid(x,a)->float: return a*x[0]**2 + x[1]**2
LB = np.array([-1.,-1.])
UB = np.array([1.,1.])
```

```python
min = sa(paraboloid,LB,UB,tau = 1e3, T0 = 100, args=[1.],min = True)
```

Example of the interpolating polynomial function

```python
import matplotlib.pyplot as plt
from InterpolatingPolynomials import interpolatingPolynomial
from numpy.random import random
from numpy import linspace

data_x = random(10)
data_y = random(10)

poly = interpolatingPolynomial(data_x,data_y)

x = linspace(0,1,100)
plt.plot(x,poly(x))
plt.scatter(data_x,data_y)
#plt.ylim(0,1)
plt.show()
```

```python
from InterpolatingPolynomials import interpolatingPolynomial_derivative
poly_d = interpolatingPolynomial_derivative(data_x,data_y)

x = linspace(0,1,100)
plt.plot(x,poly_d(x))
#plt.scatter(data_x,data_y)
#plt.ylim(0,1)
plt.show()
```
